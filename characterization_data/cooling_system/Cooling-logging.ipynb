```python
def plot_timeseries(plotting_dict):
    plt.clf()
    sns.set_theme(style="darkgrid")
    sns.set_style(rc = {'axes.facecolor':'#F0EBEB'})
    sns.set_context('talk')
    timeseries = sns.lineplot(data=plotting_dict['data'],x=plotting_dict['x'],y=plotting_dict['y'],
                             color='k',lw=2)
    timeseries.axhline(1, ls='--',lw=2,c='b')
    timeseries.axhline(4, ls='--',lw=2,c='r')
    timeseries.set_xlabel('Time (s)')
    timeseries.set_ylabel('Temperature ($^{\circ}$C)')
    plt.savefig('cooling-system_characterization-plot.jpg',dpi=300,bbox_inches = "tight")
```

```python
%matplotlib inline
plot_params = {
    'data': df,
    'x' : 'time',
    'y': 'temperature'
}
plot_timeseries(plot_params)
```
